### 2. Rerunning Dendrogram code with 1980 PCA Dataset

```python
distance_single = linkage(PCA1980.loc[:,
['PCA1','PCA2','PCA3','PCA4','PCA5',
'PCA6','PCA7','PCA8','PCA9']],method="single")
plt.figure(figsize=(18,6))
dendrogram(distance_single, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Single Method",fontsize=18)
plt.show()
```

```python
distance_single = linkage(PCA1980.loc[:,
['PCA1','PCA2','PCA3','PCA4','PCA5',
'PCA6','PCA7','PCA8','PCA9']],method="complete")
plt.figure(figsize=(18,6))
dendrogram(distance_single, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Complete Method",fontsize=18)
plt.show()
```

```python
distance_single = linkage(PCA1980.loc[:,
['PCA1','PCA2','PCA3','PCA4','PCA5',
'PCA6','PCA7','PCA8','PCA9']],method="average")
plt.figure(figsize=(18,6))
dendrogram(distance_single, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Average Method",fontsize=18)
plt.show()
```

```python
distance_single = linkage(PCA1980.loc[:,
['PCA1','PCA2','PCA3','PCA4','PCA5',
'PCA6','PCA7','PCA8','PCA9']],method="ward")
plt.figure(figsize=(18,6))
dendrogram(distance_single, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Ward Method",fontsize=18)
plt.show()
```
